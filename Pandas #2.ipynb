```
import pandas as pd
df=pd.read_csv('googleplaystore.csv')
dfcopy=df.copy()
```

```
df.Category.unique()
```

```
#первый способ без цикла
df.groupby(['Category']).App.count()
```

```
#второй способ, c помощью цикла
arr_cat=df.Category.unique()
for categor in arr_cat:
  print('{:>30} -- {}'.format(categor,df[(df['Category']==categor)].App.count()))
```

```
df.sort_values('Rating',na_position='first')
```

```
df=df[(df['Rating']>=3) & (df['Rating']<=5)]
freeapp=(df['Type']=='Free').sum()
paidapp=(df['Type']=='Paid').sum()
print('число бесплатных программ:',freeapp,';  число платных программ:',paidapp)
df
```

```
df.index=range(df.shape[0])
df
```

```
df2=dfcopy
#df2.isna.count/df2.size
print(round(df2.isna().sum()/df2.shape[0]*100,1),end='\n\n') # если имеется в виду до 2х значазих цифр
print(round(df2.isna().sum()/df2.shape[0]*100,2),end='\n\n') #если имеется до 2х десятичных знаков
print(round(df2.isna().sum().sum()/(df2.size-df2.isna().sum().sum())*100,2))
```
