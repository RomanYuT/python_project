### Задание 4

```
import os
import pandas as pd
import numpy as np

os.mkdir('bookings')
```

```
# загружаю вручную в папку bookings все файлы и после сортирую файлы по номерам:
filelist= os.listdir('bookings')
filelist.sort()
filelist.append(filelist.pop(1))
print(filelist)
```

```
dflist=[]
for filename in filelist: #загружаю данные из файлов в список из массивов
    dflist.append(pd.read_csv('bookings/'+filename, index_col = 0))
df=pd.concat(dflist) #соединяю массивы в один
df
```

```
df.shape
```

```
import matplotlib.pyplot as plt # Загружаем модуль matplotlib.pyplot
import seaborn as sns # Загружаем модуль seaborn
%matplotlib inline
fig, ax = plt.subplots(figsize=(20,12))  #Создаем область под график
sns_heatmap = sns.heatmap(df.isna(), yticklabels=False, cbar=False, cmap='viridis') # Визуализируем прпуски
plt.show()
```

```
round(df.isna().sum() / df.shape[0], 3) #доля пропусков
```

```
dfdro=df.dropna(axis=0, how='all'); 
texttodel =df.iloc[1,3] #получаем значение 'For notes:'
#ищем строки только с 'For notes:' и удаляем их:
dfdro=dfdro[~((df.hotel==texttodel) & (df.isna().sum(axis=1)==35))]; 
dfdro=dfdro.dropna(axis=1, how='all')
dfdro.fillna('unknown',inplace=True)
dfdro
```

```
dfdro.groupby(['hotel', 'arrival_date_year','arrival_date_month']).count().lead_time
```
